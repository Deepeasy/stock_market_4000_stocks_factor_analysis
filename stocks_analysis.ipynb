```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time

# Initialize ChromeDriver with webdriver-manager
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.implicitly_wait(3)

# List of stock symbols
#symbol_sgx = ['ADCINDIA.BO', 'GLOBOFFS.BO', 'GOODRICKE.BO']  # Add more symbols as needed

# List to hold data for all stocks and to track stocks with missing data
all_stock_data = []
missing_data_stocks = []

# Loop through each stock symbol
for stock_symbol in symbol_sgx:
    # Define the URL of the stock page
    url = f"https://finance.yahoo.com/quote/{stock_symbol}/key-statistics"
    driver.get(url)
    time.sleep(3)  # Allow time for the page to load

    # Dictionary to hold data for the current stock
    table_data_dict = {"Stock": stock_symbol}

    # Extract table data with first column labels and second column values
    try:
        table_rows = driver.find_elements(By.XPATH, '//tr[td[contains(@class, "yf-kbx2lo")]]')
        
        # Check if data was found
        if not table_rows:
            missing_data_stocks.append(stock_symbol)
            print(f"No data found for {stock_symbol}")
            continue  # Skip to the next stock symbol

        for row in table_rows:
            columns = row.find_elements(By.TAG_NAME, "td")
            if len(columns) >= 2:  # Ensure there are at least two columns
                label = columns[0].text  # First column text
                value = columns[1].text  # Second column text
                table_data_dict[label] = value

    except Exception as e:
        missing_data_stocks.append(stock_symbol)
        print(f"Could not extract data for {stock_symbol}: {e}")
        continue

    # Append the dictionary to the list of all stock data
    all_stock_data.append(table_data_dict)

# Close the driver
driver.quit()

# Convert the list of dictionaries into a DataFrame
stocks_df = pd.DataFrame(all_stock_data)

# Display the DataFrame with data for all stocks
print("All Stocks Data:")
print(stocks_df)

# Display the list of stocks with missing data
print("\nStocks with missing data:")
print(missing_data_stocks)
```

```python
def calculate_RSI(ticker, period='1mo', interval='1d', window_length=14):
    """
    Calculate the Relative Strength Index (RSI) for a given stock ticker.

    :param ticker: Stock ticker symbol
    :param period: Period to fetch data for (default '1mo' - 1 month)
    :param interval: Data interval (default '1d' - daily)
    :param window_length: Window length for RSI calculation (default 14)
    :return: RSI value for the last day in the fetched data
    """
    # Fetch historical data from Yahoo Finance
    data = yf.download(ticker, period=period, interval=interval)
    
    # Ensure data is fetched successfully
    if data.empty:
        raise ValueError(f"No data found for ticker {ticker}")

    # Calculate the price changes
    delta = data['Close'].diff()

    # Separate gains and losses
    gain = (delta.where(delta > 0, 0)).fillna(0)
    loss = (-delta.where(delta < 0, 0)).fillna(0)

    # Calculate the average gain and average loss
    avg_gain = gain.rolling(window=window_length, min_periods=1).mean()
    avg_loss = loss.rolling(window=window_length, min_periods=1).mean()

    # Calculate the Relative Strength (RS)
    rs = avg_gain / avg_loss

    # Calculate the RSI
    rsi = 100 - (100 / (1 + rs))

    # Get the last RSI value
    last_rsi = rsi.iloc[-1]

    return last_rsi
```

```python
def calculate_SMA(ticker, window_length, period='6mo', interval='1d'):
    """
    Calculate the Simple Moving Average (SMA) for a given stock ticker and window length.

    :param ticker: Stock ticker symbol
    :param window_length: Window length for SMA calculation (e.g., 20 or 50)
    :param period: Period to fetch data for (default '6mo' - 6 months)
    :param interval: Data interval (default '1d' - daily)
    :return: SMA value for the last day in the fetched data
    """
    # Fetch historical data from Yahoo Finance
    data = yf.download(ticker, period=period, interval=interval)
    
    # Ensure data is fetched successfully
    if data.empty:
        raise ValueError(f"No data found for ticker {ticker}")

    # Calculate the moving average
    sma = data['Close'].rolling(window=window_length).mean()
    
    # Get the last moving average value
    last_sma = sma.iloc[-1]
    
    return last_sma
```

```python
def Status(row):    
    if row["RSI"] > 70 and row["SMA_20"] > row["SMA_50"]: 
        return 1.0
    elif row["RSI"] < 30 and row["SMA_20"] < row["SMA_50"]: 
        return -1.0
    else:
        return 0.0
              
```

```python
import yfinance as yf
data_list = []

# Calculate RSI and SMA for each ticker and store in list
for ticker in symbol_sgx:
    try:
        rsi_value = calculate_RSI(ticker)
        sma_20_value = calculate_SMA(ticker, window_length=20)
        sma_50_value = calculate_SMA(ticker, window_length=50)
        data_list.append({'Ticker': ticker, 'RSI': float(rsi_value), 'SMA_20': float(sma_20_value), 'SMA_50': float(sma_50_value)})
    except ValueError as e:
        print(f"Error fetching data for {ticker}: {e}")

# Convert list to DataFrame
data_values = pd.DataFrame(data_list)

# Display the RSI and SMA values in a table
print(data_values)
```
